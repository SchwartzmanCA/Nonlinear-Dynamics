## Question 1

Write a program that steps through a data file and embeds the time-series data, producing the corresponding trajectory in reconstruction space.  This program should take a time series that is evenly sampled at some interval $\Delta t$, a time delay $\tau$ that is an integer number of multiples of $\Delta t$, and an embedding dimension $m$.  It should produce an array (list) of $m$-dimensional vectors (points in reconstruction space) each of whose $i^{th}$ element is $x(t - i\tau)$ for $i = 0 \dots m - 1$; *i.e.*, a list of vectors of the form $$[x(t), x(t-\tau), . . ., x(t-(m-1)\tau)].$$ Finally, it should plot a two-dimensional projection of this embedding.

```python
datafile = open("amplitude.dat", "r")
timeseries = []
for y in datafile:
    timeseries.append(float(y))
datafile.close()

def embed(timeseries, tau, m):
    embedding = []
    vector = []
    for t in range(len(timeseries) - m*tau):
        vector.clear()
        for j in range(m+1):
            vector.append(timeseries[(m*tau + t) - (j*tau)])
        embedding.append(vector.copy())
    return embedding
```

## Question 1.a

Run your embedding program on the **amplitude.dat** time series with $\tau = 8$ and $m = 7$.  Plot the zeroth element of the reconstructed state vector -- $x(t)$ -- on the horizontal axis and the second ($x(t-2\tau)$, here) on the vertical axis (i.e., $j = 0$ and $k = 2$).

```python
embedding = embed(timeseries, 8, 7)

x_points = []
for i in embedding:
    x_points.append(i[0])
y_points = []
for i in embedding:
    y_points.append(i[2])
    
plt.plot(x_points, y_points, 'r,')
plt.show()
```
