## Question 1

Write a simple program that uses a loop to compute a specified number $n$ of iterates of the logistic map, starting from a specified initial condition $x_0$, with a specified $r$ parameter.  Your program should take $n$, $x_0$, and $r$ as inputs or arguments.

```python
def logistic_map(r, x_0, n):
    x = [x_0]
    for i in range(1,n+1):
        x.append(r*x[i-1]*(1-x[i-1]))
    return x
```

## Question 2

Calculate the third iterate of the logistic map with $r = 2.5$ starting from $x_0 = 0.5$.

```python
print(logistic_map(2.5,0.5,3))
```

## Question 1

Use your logistic map program to explore the idea of *sensitive dependence on initial conditions*. Generate two trajectories of the logistic map ${x_n}$ and ${\hat{x}_n}$. For each of these experiments the initial condition of ${x_n}$ should be $x_0 = 0.2$, and for ${\hat{x}_n}$ you should use $x_0 = 0.200001$.  For now, fix $r = 2$, generate these two trajectories, and then plot $\lvert x_n - \hat{x}_n \rvert$ vs $n$.

```python
trajectory_a = logistic_map(2,0.2,200)
trajectory_ahat = logistic_map(2,0.200001,200)
trajectory_adifference = []
for i in range(len(trajectory_a)):
    trajectory_adifference.append(abs(trajectory_a[i] - trajectory_ahat[i]))
plt.plot(trajectory_adifference,'b.')
plt.xlabel("$n$")
plt.ylabel("$x_n$")
plt.title("$r = 2$")
plt.show()
```

## Question 1.g

For $r = 3.72$, what is the average absolute difference for the first 5,000 iterates?  (i.e., $\displaystyle \frac{1}{5,000}\sum_j^{5,000} \lvert x_j - \hat{x}_j \rvert$) 

```python
trajectory_g = logistic_map(3.72,0.2,5000)
trajectory_ghat = logistic_map(3.72,0.20001,5000)
trajectory_gdifference = []
for i in range(len(trajectory_g)):
    trajectory_gdifference.append(abs(trajectory_g[i] - trajectory_ghat[i]))

avg_gdiff = 0
for i in range(len(trajectory_gdifference)):
    avg_gdiff += trajectory_gdifference[i]
avg_gdiff /= 5000

print(avg_gdiff)
```
