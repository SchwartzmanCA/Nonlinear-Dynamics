```python
import matplotlib.pyplot as plt
import numpy as np
plt.ioff()

def logistic_map(r, x_0, n):
    x = [x_0]
    for i in range(1,n+1):
        x.append(r*x[i-1]*(1-x[i-1]))
    return x

def logistic_bifurcation(x_0, r_min, r_max, r_step, n, k):
    points = []
    for i in np.arange(r_min,r_max,r_step):
        r_i = logistic_map(i, x_0, n)
        r_i = r_i[k:]
        points.append(r_i)
    plt.plot(np.arange(r_min,r_max,r_step),points,'.')
    plt.rcParams['figure.figsize'] = [16, 12]
    plt.show()
```

```python
logistic_bifurcation(0.2,2.4,4,0.01,1000,5)
```

```python
def general_map(f, r, x_0, n):
    x = [x_0]
    for i in range(1,n+1):
        x.append(f(r,x[i-1]))
    return x

def general_bifurcation(f, x_0, r_min, r_max, r_step, n, k):
    points = []
    for i in np.arange(r_min,r_max,r_step):
        r_i = general_map(f, i, x_0, n)
        r_i = r_i[k:]
        points.append(r_i)
    plt.plot(np.arange(r_min,r_max,r_step),points,',k', alpha=0.1)
    plt.rcParams['figure.figsize'] = [16, 12]
    plt.show()
```

```python
general_bifurcation(lambda r,x: r - x*x, 0.2, 0.1, 2, 0.0001, 1000, 500)
```
