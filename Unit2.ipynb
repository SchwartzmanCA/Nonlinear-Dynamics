## Question 1

Write a program to construct a bifurcation diagram for the logistic map. (*Hint: this should be a loop that calls your logistic map program from Unit 1.3.*) Your program should take the following arguments:

+ An initial condition $x_0$
+ An $r_{min}$ and an $r_{max}$ that specify a range of $r$ values for the $x$-axis of your bifurcation plot
+ An $r_{step}$, which specifies how many "slices" your plot has
+ A number $n$ of total iterates to perform
+ A number $k$ of total iterates to discard without plotting (*i.e.*, to remove the transient)

The output should be a plot of the logistic map bifurcation diagram for a range of $r$.

```python
def logistic_map(r, x_0, n):
    x = [x_0]
    for i in range(1,n+1):
        x.append(r*x[i-1]*(1-x[i-1]))
    return x

def logistic_bifurcation(x_0, r_min, r_max, r_step, n, k):
    points = []
    for i in np.arange(r_min,r_max,r_step):
        r_i = logistic_map(i, x_0, n)
        r_i = r_i[k:]
        points.append(r_i)
    plt.plot(np.arange(r_min,r_max,r_step),points,'.')
    plt.rcParams['figure.figsize'] = [16, 12]
    plt.show()
```

#### Check your program by constructing a bifurcation plot for $r \in [2.4,4]$ with a step size of $0.01$.  For each $r$, construct a 1000-iterate trajectory from $x_0 = 0.2$, and discard the first five points -- *i.e.*, plot $x_5$ to $x_{1000}$ for each $r$.

```python
logistic_bifurcation(0.2,2.4,4,0.01,1000,5)
```

## Question 3

Here I modified the bifurcation function to be able to take any mathematical function as input, instead of just the logistic map.

```python
def general_map(f, r, x_0, n):
    x = [x_0]
    for i in range(1,n+1):
        x.append(f(r,x[i-1]))
    return x

def general_bifurcation(f, x_0, r_min, r_max, r_step, n, k):
    points = []
    for i in np.arange(r_min,r_max,r_step):
        r_i = general_map(f, i, x_0, n)
        r_i = r_i[k:]
        points.append(r_i)
    plt.plot(np.arange(r_min,r_max,r_step),points,',k', alpha=0.1)
    plt.rcParams['figure.figsize'] = [16, 12]
    plt.show()
```

#### Then I ran this generalized bifurcation function with the map $x_{n+1} = r - x^2_n$.

```python
general_bifurcation(lambda r,x: r - x*x, 0.2, 0.1, 2, 0.0001, 1000, 500)
```
